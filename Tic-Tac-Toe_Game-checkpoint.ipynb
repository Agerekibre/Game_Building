```python
# Step 1 import libraries
# Step 2 play ground 3x3 matrix
# Step 3 player assignment x or 0
# Step 4 marks the locations of players inputs in the board
# Step 5 compares if three inputs in a horizontal or vertical or diagonal sequence
# Step 6 randomly selects the player who can start the game
# Step 7 checks if there is an empty box to make a move
# Step 8 checks if the boxes are full
# Step 9 checks if the user input is a valid integer 
# Step 10 checks if the players input is an integer between 1 and 9-, and players choose a free box
# Step 11 gives an option to play the game again


#Functions defined for the game
# display_board
# player_assignment
# who_starts
# place_marker
# win_check
# isboxfree
# isboardfull
# isvalidinput
# location_choice
# replay


#Importing libraries
import random #used to randomly select which player can start playing
from IPython.display import clear_output #clear_out updates the play board after eachplayer's input

#All the functions used are defined below

#function to display the playground
def display_board(board): 
    
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
    

#This function assigns which player is X and which one is O
def player_assignment():
    marker = ""
    
    while not (marker == "X" or marker == "O"): #keeps asking for choice until X or O is chosen
        marker = input("Player : Do you want to be X or O)").upper()
        
    if marker == "X":
        return ("X","O")
    else:
        return ("O","X")

    
#Randomly selects which player starts the game
def who_starts():
    if random.randint(0, 1) == 0: 
        return 'O'
    else:
        return 'X'

    
#Marks the position of the players input in the board
def place_marker(board, marker, position):
    clear_output() #after every player's move the board is updated with the new player input
    board[position] = marker #marker is marked as players sign in the specific position of the board
    
    
#Checks the winning positions to identify the winner
def win_check(board,mark):
    
    return ((board[1] == mark and board[2] == mark and board[3] == mark) or # Horizontal
            (board[4] == mark and board[5] == mark and board[6] == mark) or # Horizontal
            (board[7] == mark and board[8] == mark and board[9] == mark) or # Horizontal 
            (board[7] == mark and board[4] == mark and board[1] == mark) or # Vertical
            (board[8] == mark and board[5] == mark and board[2] == mark) or # Vertical
            (board[9] == mark and board[6] == mark and board[3] == mark) or # Vertical
            (board[7] == mark and board[5] == mark and board[3] == mark) or # Right Diangnoal
            (board[9] == mark and board[5] == mark and board[1] == mark)) # Left Diangoal    
    

#Checks if the box is free to make a move
def isboxfree(board, move):
    return board[move] == ' ' # returns True value if a box in the board is free to make a move 


#Checks if the board is full or not
def isboardfull(board):
    for i in range(1,10):
        if isboxfree(board, i):#calls isboxfree function to decide the isboardfull condition,if free box,then returns False
            return False
    return True


#Evaluates the input from players; incase of wrong input, it keeps asking for the correct input from the players
def isvalidinput():
    while True:
        position = input('Choose your next position: (1-9) ')
        if position.isnumeric()== True:#checks if the used position marker is a numberic value
            break
        else:
            continue
    return int(position)


'''Checks if the input from the players is an integer between 1 and 9
   and also checks if the players choose an empty box and keeps asking
   for the proper input and choice of a free box'''
def location_choice(board):
    position = 0
    while position not in [1,2,3,4,5,6,7,8,9] or not isboxfree(board, position):
        position = isvalidinput()  
    return position    


# Asking the players if they want to continue again
def replay():
    playagain = input("Do you want to play again? type y for yes and n for no ")
    if playagain.lower() == 'y':
        print("Thank you for choosing to play again!")
        return True
    elif playagain.lower() == 'n':
        print("Thank you for playing.")
        return False
    else:
        return replay() #Continues asking for the replay unless y or n is typed


#Logic part of the code

#This while loop continues until the game is finished and players do not want to play again
while True:
    theboard = ["#"," "," "," "," "," "," "," "," "," "] #prints the playing board
    player1_marker , player2_marker = player_assignment() #the players' merking signs are assigned  
    starter = who_starts() #starter of the game is randomly chosen
    print(starter + " will start playing")
    
    
    if player1_marker or player2_marker == "X" or "O": #Game will start after a player choses playng mark
        game_on = True
    else:
        game_on = False

    while game_on: #the game starts for True value of sign choice from X and O
        
        if starter == "X":
            
            display_board(theboard)
            position = location_choice(theboard) #position is marked on the chosen location of the board
            place_marker(theboard,'X', position)
            
            if win_check(theboard, 'X'): #winning condition is checked for player X by calling the function win_check
                display_board(theboard)
                print('Congratulations! X wins the game!')
                game_on = False
            else:
                if isboardfull(theboard): #checks if the board is full without having a winner
                    display_board(theboard)
                    print('The game is a draw!')
                    break
                else:
                    starter = 'O'

        else:
            display_board(theboard)
            position = location_choice(theboard)
            place_marker(theboard, 'O', position)

            if win_check(theboard, 'O'):
                display_board(theboard)
                print('Congratulations! O wins the game!')
                game_on = False
            else:
                if isboardfull(theboard):
                    display_board(theboard)
                    print('The game is a draw!')
                    break
                else:
                    starter = 'X'
                    
    if not replay(): #chacks the replay answer from the players
        break
```
